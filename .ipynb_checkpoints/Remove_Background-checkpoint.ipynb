```python
from PIL import Image
import numpy as np
import os
```

```python
# make background transparent 
def remove_background(directory, filename, new_directory):
    img_file_path = directory + '/' + filename
    img = Image.open(img_file_path)
    img = img.convert("RGBA")
    datas = img.getdata()

    newData = []
    for item in datas:
        if item[0] >= 245 and item[1] >= 245 and item[2] >= 245:
            newData.append((255, 255, 255, 0))
        else:
            newData.append(item)
            
    img.putdata(newData)
    new_path = new_directory + '/' + filename
    img.save(new_path, "PNG")
```

```python
directories = ['Set 2', 'Set 3', 'Set 4', 'Set 5', 'Set 6']
new_directories = ['Set_2_cleaned', 'Set_3_cleaned', 'Set_4_cleaned', 'Set_5_cleaned', 'Set_6_cleaned']
for directory, new_directory in zip(directories, new_directories):
    for filename in os.listdir(directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            remove_background(directory, filename, new_directory)
            print(directory + '/' + filename)
```
